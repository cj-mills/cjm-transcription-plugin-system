```python
# Shim verification: class-identity + a minimal save/get_cached cycle
# (note the FOUR-key cache identity: audio, audio_hash, text_hash,
# config_hash — the task input is the (audio, transcript) pair).
import tempfile, os
import cjm_forced_alignment_adapter_interface.storage as _new

assert ForcedAlignmentRow is _new.ForcedAlignmentRow
assert ForcedAlignmentStorage is _new.ForcedAlignmentStorage

with tempfile.TemporaryDirectory() as td:
    db = os.path.join(td, 'fa.db')
    st = ForcedAlignmentStorage(db)
    st.save(job_id='j1', audio_path='/a.wav', audio_hash='sha256:aa',
            text='hello world', text_hash='sha256:tt', config_hash='sha256:cc',
            items=[{'text': 'hello', 'start_time': 0.0, 'end_time': 0.5}])
    row = st.get_cached('/a.wav', 'sha256:aa', 'sha256:tt', 'sha256:cc')
    assert row is not None and row.items[0]['text'] == 'hello'
print('forced_alignment_storage shim OK')
```
